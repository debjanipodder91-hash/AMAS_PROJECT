```python
import numpy as np
from scipy.stats import genextreme
import pandas as pd

"""rewriting data to file without leading # and without spaces and overwrite existing file"""

filename=f'fitted_files.csv'
data = []

with open(filename, "r") as f:
    for line in f:
        line = line.strip()  # Remove leading/trailing whitespace
        if line.startswith("#"):
            line = line[1:]  # Remove leading '#'
        line = line.replace(" ", "")  # Remove all spaces
        print(line)
        data.append(line)


```

```python
with open(filename, "w") as f:
    for x in data:
        f.write(f"{x}\n")

```
