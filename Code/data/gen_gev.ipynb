```python
import numpy as np
from scipy.stats import genextreme
import pandas as pd

# Generated Data

N = 1_000#e6

# Dataset parameters
mu    = 70
alpha = 10
gamma = 0.3
data = genextreme.rvs(c=-gamma, loc=mu, scale=alpha, size=N, random_state=42)

filename=f'gev-mu={mu},alpha={alpha},gamma={gamma},N=e3c.csv'

with open(filename, "w") as f:
    for x in data:
        f.write(f"{x}\n")

with open("generated_files.csv", "a") as f:
    f.write(f"{filename}\n")      

print(f"File created: {filename} and saved to generated_files.csv")

# Load Data
df   = pd.read_csv(filename)

data = df.values
data
```
