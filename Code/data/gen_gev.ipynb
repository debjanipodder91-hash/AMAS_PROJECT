```python
import numpy as np
from scipy.stats import genextreme
import pandas as pd

# Generated Data

N = 100000

# Dataset parameters
mu    = 7
alpha = 200
gamma = -0.5
data = genextreme.rvs(c=-gamma, loc=mu, scale=alpha, size=N, random_state=42)

filename=f'gev-mu={mu},alpha={alpha},gamma={gamma},N={N}.csv'

with open(filename, "w") as f:
    for x in data:
        f.write(f"{x}\n")

print(f"File created: {filename}")

# Load Data
df   = pd.read_csv(filename)

data = df.values
data
```
