Generating Extreme Event data for AMAS Project

from gaussian distribution

interested in the upper tail extremes

data is cut away 5 sigma from origin

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from matplotlib.markers import MarkerStyle

my_cmap = plt.colormaps["turbo"]
my_marker=MarkerStyle(marker="1", fillstyle='full', transform=None, capstyle=None, joinstyle=None)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['savefig.dpi']=600
plt.rcParams["savefig.bbox"]='tight'
plt.rcParams["figure.figsize"]='5,3'
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontfamily='monospace')

r = np.random
r.seed(42)
```

```python
"""dataset parameters"""
filename="gauss-mu=5,sigma=2.0,N=e9,cutoff=4sig.csv"
mu=5
sigma=2.0
N=1_000_000_000
sigma_cutoff=4

print(f"{N:.1e}")
extreme_values = []
```

```python
divider = 1_000
for i in range(divider):
    print(f"Processing chunk {i+1}/{divider}...", end="\r")
    x_array = r.normal(loc=mu, scale=sigma, size=N//divider)
    mask = x_array - mu >= sigma_cutoff * sigma
    
    extreme_values.append(x_array[mask])
#x_array[mask]
```

```python
extreme_values = np.concatenate(extreme_values)

"""write the IDs of the predicted high and low earners to separate text files"""
with open(filename, "w") as f:
    for x in extreme_values:
        f.write(f"{x}\n")


"""read the values from the text file"""
extreme_values_from_file = []
with open(filename, "r") as f:
    for line in f:
        extreme_values_from_file.append(float(line.strip()))

print(f"Number of extreme values: {len(extreme_values_from_file)}")
```
